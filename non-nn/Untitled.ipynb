```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python

# Step 1: Load training data
train_df = pd.read_csv('../sim_og_norm.csv')
```

```python
X = train_df[['a', 'q', 'delta_e']].values
y = train_df[['Cd', 'Cl', 'Cm']].values
```

```python
# Step 2: Train linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Step 3: Predict on training data for evaluation
y_pred = model.predict(X)
```

```python
# Evaluation metrics
r2 = r2_score(y, y_pred, multioutput='raw_values')
mse = mean_squared_error(y, y_pred, multioutput='raw_values')
```

```python
print(f"R² scores: y1={r2[0]:}, y2={r2[1]:}, y3={r2[2]:}")
print(f"MSE values: y1={mse[0]:}, y2={mse[1]:}, y3={mse[2]:}")
```

```python

# Step 4: Load new data to predict
predict_df = pd.read_csv("../normalized_output_test.csv")
X_new = predict_df[['Alpha', 'q', 'delta_e']]
y_new_pred = model.predict(X_new)
```

```python
# Step 5: Save predictions to CSV
pred_df = pd.DataFrame(y_new_pred, columns=['pred_y1', 'pred_y2', 'pred_y3'])
output_df = pd.concat([predict_df, pred_df], axis=1)
output_df.to_csv('predicted_output.csv', index=False)

print("Predictions saved to predicted_output.csv")
```
